```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import uuid

classic_stats = [
    'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
    'OREB', 'DREB', 'STL', 'BLK', 'TO', 'PF'
]

four_factors = ['eFG%', 'TO%', 'OR%', 'FTR']

reduced_stats = [
    'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGA', 'FG3M', 'FTM', 'DREB', 'STL', 'BLK', 'TO',
    'eFG%', 'TO%', 'OR%', 'FTR', 'ELO_HOME_BEFORE', 'ELO_AWAY_BEFORE'
]

feature_sets = {
    'classic': classic_stats,
    'four_factors': four_factors,
    'classic_plus_four': classic_stats + four_factors,
    'reduced': reduced_stats
}

all_reg_predictions = []

for set_name, feature_list in feature_sets.items():
    print(f"\n Evaluating Feature Set: {set_name.upper()} ")
    results = []

    for x in list(range(1, 21)) + [21]:
        print(f"\n>> Rolling window: {x}")

        try:
            if x == 21:
                df_source = pd.read_csv("../df_elo_season_averages.csv")
                prefix_type = "AVG_SEASON"
            else:
                df_source = pd.read_csv(f"../Rolling_averages_sets/df_elo_{x}.csv")
                prefix_type = f"AVG_LAST_{x}"
        except FileNotFoundError:
            print(f"Missing file for x={x}. Skipping.")
            continue

        df_source["POINT_DIFF"] = df_source["PTS_HOME"] - df_source["PTS_AWAY"]

        for train_season in range(2004, 2019):
            test_season = train_season + 1

            train_df = df_source[df_source["SEASON"] == train_season]
            test_df = df_source[df_source["SEASON"] == test_season]

            if train_df.empty or test_df.empty:
                continue

            rolling_features = [stat for stat in feature_list if 'ELO' not in stat]
            elo_features = [stat for stat in feature_list if 'ELO' in stat]

            feature_cols = [f"HOME_{stat}_{prefix_type}" for stat in rolling_features] + \
                           [f"AWAY_{stat}_{prefix_type}" for stat in rolling_features] + \
                           elo_features

            X_train = train_df[feature_cols]
            y_train = train_df["POINT_DIFF"]
            X_test = test_df[feature_cols]
            y_test = test_df["POINT_DIFF"]

            pipeline = Pipeline([
                ('scaler', StandardScaler()),
                ('linreg', LinearRegression())
            ])

            pipeline.fit(X_train, y_train)
            y_pred = pipeline.predict(X_test)

            test_df = test_df.copy()
            test_df["POINT_DIFF"] = test_df["PTS_HOME"] - test_df["PTS_AWAY"]
            test_df["PREDICTED_POINT_DIFF"] = y_pred
            test_df["PREDICTED_HOME_TEAM_WINS"] = (y_pred >= 0).astype(int)

            accuracy = (test_df["PREDICTED_HOME_TEAM_WINS"] == test_df["HOME_TEAM_WINS"]).mean()
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            r2 = r2_score(y_test, y_pred)

            results.append({
                'rolling_window': x,
                'train_season': train_season,
                'test_season': test_season,
                'regression_rmse': rmse,
                'regression_r2': r2,
                'accuracy_from_regression': accuracy
            })

            print(f"Train {train_season} → Test {test_season} | x={x} | Acc: {accuracy:.4f} | RMSE: {rmse:.2f}")

            test_df["POINT_DIFF"] = test_df["PTS_HOME"] - test_df["PTS_AWAY"]

            game_info = test_df[[
                "GAME_ID", "SEASON", "TEAM_ID_HOME", "TEAM_ID_AWAY", "HOME_TEAM_WINS", "POINT_DIFF"
            ]].copy()
            game_info["PREDICTED_POINT_DIFF"] = y_pred
            game_info["PREDICTED_HOME_TEAM_WINS"] = (y_pred >= 0).astype(int)
            game_info["MODEL"] = "linear_regression"
            game_info["TASK"] = "regression"
            game_info["DATASET"] = set_name
            game_info["ROLLING_WINDOW"] = x
            game_info["TEST_SEASON"] = test_season
            game_info["EXPERIMENT_ID"] = str(uuid.uuid4())

            all_reg_predictions.append(game_info)

    df_results = pd.DataFrame(results)
    df_results.to_csv(f"results2_linreg_{set_name}.csv", index=False)
    print(f"\n✅ Results saved: results2_linreg_{set_name}.csv")

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=df_results,
        x='rolling_window',
        y='accuracy_from_regression',
        hue='test_season',
        palette='tab10'
    )
    plt.title(f'Linear Regression Accuracy from Predicted Point Diff — {set_name.upper()}', fontsize=14)
    plt.xlabel('Rolling Window Size (1–20), 21 = Season-to-Date Average', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.xticks(range(1, 22))
    plt.grid(True)
    plt.legend(title='Test Season', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

all_preds_df = pd.concat(all_reg_predictions, ignore_index=True)
all_preds_df.to_csv("all_game_predictions_linear_regression.csv", index=False)
print("✅ All per-game regression predictions saved: all_game_predictions_linear_regression.csv")
```
