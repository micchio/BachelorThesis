```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, log_loss, confusion_matrix
)
import matplotlib.pyplot as plt
import seaborn as sns
import uuid

classic_stats = [
    'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
    'OREB', 'DREB', 'STL', 'BLK', 'TO', 'PF'
]

four_factors = ['eFG%', 'TO%', 'OR%', 'FTR']

reduced_stats = [
    'PTS', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGA', 'FG3A', 'DREB', 'STL', 'BLK',
    'eFG%', 'TO%', 'OR%', 'FTR',
    'ELO_HOME_BEFORE', 'ELO_AWAY_BEFORE'
]

feature_sets = {
    'classic': classic_stats,
    'four_factors': four_factors,
    'classic_plus_four': classic_stats + four_factors,
    'reduced': reduced_stats
}

model_name = "naive_bayes"
task = "classification"
all_class_predictions = []

for set_name, feature_list in feature_sets.items():
    print(f"\n Evaluating Feature Set: {set_name.upper()} ")
    results = []

    for x in list(range(1, 21)) + [21]:
        print(f"\n>> Rolling window: {x}")

        try:
            if x == 21:
                df_source = pd.read_csv("../df_elo_season_averages.csv")
                prefix_type = "AVG_SEASON"
            else:
                df_source = pd.read_csv(f"../Rolling_averages_sets/df_elo_{x}.csv")
                prefix_type = f"AVG_LAST_{x}"
        except FileNotFoundError:
            print(f"Missing file for x={x}. Skipping.")
            continue

        for train_season in range(2004, 2019):
            test_season = train_season + 1

            train_df = df_source[df_source['SEASON'] == train_season]
            test_df = df_source[df_source['SEASON'] == test_season]

            if train_df.empty or test_df.empty:
                continue

            test_df = test_df.copy()
            test_df["POINT_DIFF"] = test_df["PTS_HOME"] - test_df["PTS_AWAY"]

            rolling_features = [stat for stat in feature_list if 'ELO' not in stat]
            elo_features = [stat for stat in feature_list if 'ELO' in stat]

            feature_cols = [f"HOME_{stat}_{prefix_type}" for stat in rolling_features] + \
                           [f"AWAY_{stat}_{prefix_type}" for stat in rolling_features] + \
                           elo_features

            X_train = train_df[feature_cols]
            y_train = train_df['HOME_TEAM_WINS']
            X_test = test_df[feature_cols]
            y_test = test_df['HOME_TEAM_WINS']

            model = GaussianNB()
            tscv = TimeSeriesSplit(n_splits=10)
            cv_scores = cross_val_score(model, X_train, y_train, cv=tscv, scoring='accuracy')

            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            y_prob = model.predict_proba(X_test)[:, 1]

            acc = accuracy_score(y_test, y_pred)
            prec = precision_score(y_test, y_pred, zero_division=0)
            rec = recall_score(y_test, y_pred, zero_division=0)
            f1 = f1_score(y_test, y_pred, zero_division=0)
            try:
                ll = log_loss(y_test, y_prob)
            except:
                ll = None
            tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

            results.append({
                'rolling_window': x,
                'train_season': train_season,
                'test_season': test_season,
                'cv_accuracy_mean': cv_scores.mean(),
                'test_accuracy': acc,
                'precision': prec,
                'recall': rec,
                'f1_score': f1,
                'log_loss': ll,
                'true_negatives': tn,
                'false_positives': fp,
                'false_negatives': fn,
                'true_positives': tp
            })

            print(f"Train {train_season} → Test {test_season} | Acc: {acc:.4f} | F1: {f1:.4f}")

            game_info = test_df[["GAME_ID", "SEASON", "TEAM_ID_HOME", "TEAM_ID_AWAY", "HOME_TEAM_WINS", "POINT_DIFF"]].copy()
            game_info["PRED_CLASS"] = y_pred
            game_info["PROB_CLASS"] = y_prob
            game_info["MODEL"] = model_name
            game_info["TASK"] = task
            game_info["DATASET"] = set_name
            game_info["ROLLING_WINDOW"] = x
            game_info["TEST_SEASON"] = test_season
            game_info["TRAIN_SEASON"] = train_season  
            game_info["EXPERIMENT_ID"] = str(uuid.uuid4())

            all_class_predictions.append(game_info)

    df_results = pd.DataFrame(results)
    df_results.to_csv(f"Results_nb_{set_name}.csv", index=False)  
    print(f"\n✅ Results saved: Results_nb_{set_name}.csv")

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=df_results,
        x='rolling_window',
        y='test_accuracy',
        hue='test_season',
        palette='tab10'
    )
    plt.title(f'Naive Bayes Accuracy vs Rolling Window Size — {set_name.upper()}', fontsize=14)
    plt.xlabel('Rolling Window Size (1–20), 21 = Season-to-Date Average', fontsize=12)
    plt.ylabel('Test Accuracy', fontsize=12)
    plt.xticks(range(1, 22))
    plt.grid(True)
    plt.legend(title='Test Season', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

df_all_class_preds = pd.concat(all_class_predictions, ignore_index=True)
df_all_class_preds.to_csv("All_class_predictions_naive_bayes.csv", index=False)
print("✅ All per-game classification predictions saved: all_class_predictions_naive_bayes.csv")
```
