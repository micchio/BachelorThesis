```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
import uuid

classic_stats = [
    'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
    'OREB', 'DREB', 'STL', 'BLK', 'TO', 'PF'
]

four_factors = ['eFG%', 'TO%', 'OR%', 'FTR']

reduced_stats = [
    'PTS', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGA', 'FG3A', 'DREB', 'STL', 'BLK',
    'eFG%', 'TO%', 'OR%', 'FTR',
    'ELO_HOME_BEFORE', 'ELO_AWAY_BEFORE'
]

feature_sets = {
    'classic':classic_stats,
    'four_factors': four_factors,
    'classic_plus_four': classic_stats + four_factors,
    'reduced': reduced_stats
}

grid_params = {
    'svr__C': [0.1, 1, 10],
    'svr__gamma': [0.1, 1, 10],
    'svr__kernel': ['linear', 'rbf']
}

all_svr_predictions = []

for set_name, feature_list in feature_sets.items():
    print(f"\n========== Feature Set: {set_name.upper()} ==========")
    results = []

    for x in list(range(1, 21)) + [21]:
        print(f"\n>> Rolling window: {x}")

        try:
            if x == 21:
                df_source = pd.read_csv("../df_elo_season_averages.csv")
                prefix_type = "AVG_SEASON"
            else:
                df_source = pd.read_csv(f"../Rolling_averages_sets/df_elo_{x}.csv")
                prefix_type = f"AVG_LAST_{x}"
        except FileNotFoundError:
            print(f"Missing file for x={x}. Skipping.")
            continue

        df_source["POINT_DIFF"] = df_source["PTS_HOME"] - df_source["PTS_AWAY"]

        for train_season in range(2004, 2019):
            test_season = train_season + 1
            train_df = df_source[df_source["SEASON"] == train_season]
            test_df = df_source[df_source["SEASON"] == test_season]

            if train_df.empty or test_df.empty:
                continue

            test_df = test_df.copy()

            rolling_features = [stat for stat in feature_list if 'ELO' not in stat]
            elo_features = [stat for stat in feature_list if 'ELO' in stat]

            feature_cols = [f"HOME_{stat}_{prefix_type}" for stat in rolling_features] + \
                           [f"AWAY_{stat}_{prefix_type}" for stat in rolling_features] + \
                           elo_features

            X_train = train_df[feature_cols]
            y_train = train_df["POINT_DIFF"]
            X_test = test_df[feature_cols]
            y_test = test_df["POINT_DIFF"]

            pipeline = Pipeline([
                ('scaler', StandardScaler()),
                ('svr', SVR())
            ])

            grid = GridSearchCV(
                pipeline,
                grid_params,
                cv=TimeSeriesSplit(n_splits=10), 
                scoring='neg_mean_squared_error',
                n_jobs=-1,
                verbose=0
            )

            grid.fit(X_train, y_train)
            best_model = grid.best_estimator_

            y_pred = best_model.predict(X_test)
            
            y_class_pred = (y_pred >= 0).astype(int)
            
            actual_outcomes = test_df['HOME_TEAM_WINS'].values
            accuracy = (y_class_pred == actual_outcomes).mean()

            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            mae = mean_absolute_error(y_test, y_pred)
            r2 = r2_score(y_test, y_pred)

            results.append({
                'rolling_window': x,
                'train_season': train_season,
                'test_season': test_season,
                'regression_rmse': rmse,
                'regression_mae': mae,
                'regression_r2': r2,
                'classification_accuracy': accuracy,  
                'best_C': grid.best_params_['svr__C'],
                'best_gamma': grid.best_params_['svr__gamma'],
                'best_kernel': grid.best_params_['svr__kernel']
            })

            print(f"Train {train_season} → Test {test_season} | RMSE: {rmse:.2f} | R²: {r2:.4f} | Acc: {accuracy:.4f} | Best: C={grid.best_params_['svr__C']}, γ={grid.best_params_['svr__gamma']}, kernel={grid.best_params_['svr__kernel']}")

            game_info = test_df[[
                "GAME_ID", "SEASON", "TEAM_ID_HOME", "TEAM_ID_AWAY", "HOME_TEAM_WINS", "POINT_DIFF"
            ]].copy()
            game_info["PRED_POINT_DIFF"] = y_pred
            game_info["PRED_HOME_WINS"] = y_class_pred
            game_info["MODEL"] = "svr"
            game_info["TASK"] = "regression"
            game_info["DATASET"] = set_name
            game_info["ROLLING_WINDOW"] = x
            game_info["TEST_SEASON"] = test_season
            game_info["TRAIN_SEASON"] = train_season
            game_info["EXPERIMENT_ID"] = str(uuid.uuid4())

            all_svr_predictions.append(game_info)

    df_results = pd.DataFrame(results)
    df_results.to_csv(f"results_svr_regression_{set_name}.csv", index=False)
    print(f"\n✅ Results saved: results_svr_regression_{set_name}.csv")

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=df_results,
        x='rolling_window',
        y='regression_rmse',
        hue='test_season',
        palette='tab10'
    )
    plt.title(f'SVR Regression RMSE vs Rolling Window Size — {set_name.upper()}', fontsize=14)
    plt.xlabel('Rolling Window Size (1–20), 21 = Season-to-Date Average', fontsize=12)
    plt.ylabel('Test RMSE (Points)', fontsize=12)
    plt.xticks(range(1, 22))
    plt.grid(True)
    plt.legend(title='Test Season', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=df_results,
        x='rolling_window',
        y='classification_accuracy',
        hue='test_season',
        palette='tab10'
    )
    plt.title(f'SVR Classification Accuracy (from Regression) — {set_name.upper()}', fontsize=14)
    plt.xlabel('Rolling Window Size (1–20), 21 = Season-to-Date Average', fontsize=12)
    plt.ylabel('Test Accuracy', fontsize=12)
    plt.xticks(range(1, 22))
    plt.grid(True)
    plt.legend(title='Test Season', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

all_preds_df = pd.concat(all_svr_predictions, ignore_index=True)
all_preds_df.to_csv("all_regression_predictions_svr.csv", index=False)
print("✅ All per-game regression predictions saved: All_regression_predictions_svr.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
import uuid

classic_stats = [
    'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
    'OREB', 'DREB', 'STL', 'BLK', 'TO', 'PF'
]

four_factors = ['eFG%', 'TO%', 'OR%', 'FTR']

reduced_stats = [
    'PTS', 'FT_PCT', 'FG3_PCT', 'AST',
    'FGA', 'FG3A', 'DREB', 'STL', 'BLK',
    'eFG%', 'TO%', 'OR%', 'FTR',
    'ELO_HOME_BEFORE', 'ELO_AWAY_BEFORE'
]

feature_sets = {
    'classic':classic_stats,
    'four_factors': four_factors,
    'classic_plus_four': classic_stats + four_factors,
    'reduced': reduced_stats
}

grid_params = {
    'svr__C': [0.1, 1, 10],
    'svr__gamma': [0.1, 1, 10],
    'svr__kernel': ['linear', 'rbf']
}

all_svr_predictions = []

for set_name, feature_list in feature_sets.items():
    print(f"\n========== Feature Set: {set_name.upper()} ==========")
    results = []

    for x in list(range(1, 21)) + [21]:
        print(f"\n>> Rolling window: {x}")

        try:
            if x == 21:
                df_source = pd.read_csv("../df_elo_season_averages.csv")
                prefix_type = "AVG_SEASON"
            else:
                df_source = pd.read_csv(f"../Rolling_averages_sets/df_elo_{x}.csv")
                prefix_type = f"AVG_LAST_{x}"
        except FileNotFoundError:
            print(f"Missing file for x={x}. Skipping.")
            continue

        df_source["POINT_DIFF"] = df_source["PTS_HOME"] - df_source["PTS_AWAY"]

        
        for first_train_season in range(2004, 2018):  
            second_train_season = first_train_season + 1
            test_season = first_train_season + 2
            
            
            train_df_1 = df_source[df_source["SEASON"] == first_train_season]
            train_df_2 = df_source[df_source["SEASON"] == second_train_season]
            train_df = pd.concat([train_df_1, train_df_2], ignore_index=True)
            
            test_df = df_source[df_source["SEASON"] == test_season]

            if train_df.empty or test_df.empty:
                continue

            test_df = test_df.copy()

            rolling_features = [stat for stat in feature_list if 'ELO' not in stat]
            elo_features = [stat for stat in feature_list if 'ELO' in stat]

            feature_cols = [f"HOME_{stat}_{prefix_type}" for stat in rolling_features] + \
                           [f"AWAY_{stat}_{prefix_type}" for stat in rolling_features] + \
                           elo_features

            X_train = train_df[feature_cols]
            y_train = train_df["POINT_DIFF"]
            X_test = test_df[feature_cols]
            y_test = test_df["POINT_DIFF"]

            pipeline = Pipeline([
                ('scaler', StandardScaler()),
                ('svr', SVR())
            ])

            grid = GridSearchCV(
                pipeline,
                grid_params,
                cv=TimeSeriesSplit(n_splits=10), 
                scoring='neg_mean_squared_error',
                n_jobs=-1,
                verbose=0
            )

            grid.fit(X_train, y_train)
            best_model = grid.best_estimator_

            y_pred = best_model.predict(X_test)
            
            y_class_pred = (y_pred >= 0).astype(int)
            
            actual_outcomes = test_df['HOME_TEAM_WINS'].values
            accuracy = (y_class_pred == actual_outcomes).mean()

            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            mae = mean_absolute_error(y_test, y_pred)
            r2 = r2_score(y_test, y_pred)

            results.append({
                'rolling_window': x,
                'train_seasons': f"{first_train_season}-{second_train_season}",
                'test_season': test_season,
                'regression_rmse': rmse,
                'regression_mae': mae,
                'regression_r2': r2,
                'classification_accuracy': accuracy,  
                'best_C': grid.best_params_['svr__C'],
                'best_gamma': grid.best_params_['svr__gamma'],
                'best_kernel': grid.best_params_['svr__kernel']
            })

            print(f"Train {first_train_season}-{second_train_season} → Test {test_season} | RMSE: {rmse:.2f} | R²: {r2:.4f} | Acc: {accuracy:.4f} | Best: C={grid.best_params_['svr__C']}, γ={grid.best_params_['svr__gamma']}, kernel={grid.best_params_['svr__kernel']}")

            game_info = test_df[[
                "GAME_ID", "SEASON", "TEAM_ID_HOME", "TEAM_ID_AWAY", "HOME_TEAM_WINS", "POINT_DIFF"
            ]].copy()
            game_info["PRED_POINT_DIFF"] = y_pred
            game_info["PRED_HOME_WINS"] = y_class_pred
            game_info["MODEL"] = "svr"
            game_info["TASK"] = "regression"
            game_info["DATASET"] = set_name
            game_info["ROLLING_WINDOW"] = x
            game_info["TEST_SEASON"] = test_season
            game_info["TRAIN_SEASONS"] = f"{first_train_season}-{second_train_season}"
            game_info["EXPERIMENT_ID"] = str(uuid.uuid4())

            all_svr_predictions.append(game_info)

    df_results = pd.DataFrame(results)
    df_results.to_csv(f"results_svr_regression_{set_name}_train2.csv", index=False)
    print(f"\n✅ Results saved: results_svr_regression_{set_name}_train2.csv")

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=df_results,
        x='rolling_window',
        y='regression_rmse',
        hue='test_season',
        palette='tab10'
    )
    plt.title(f'SVR Regression RMSE vs Rolling Window Size (2 Train Seasons) — {set_name.upper()}', fontsize=14)
    plt.xlabel('Rolling Window Size (1–20), 21 = Season-to-Date Average', fontsize=12)
    plt.ylabel('Test RMSE (Points)', fontsize=12)
    plt.xticks(range(1, 22))
    plt.grid(True)
    plt.legend(title='Test Season', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=df_results,
        x='rolling_window',
        y='classification_accuracy',
        hue='test_season',
        palette='tab10'
    )
    plt.title(f'SVR Classification Accuracy (2 Train Seasons) — {set_name.upper()}', fontsize=14)
    plt.xlabel('Rolling Window Size (1–20), 21 = Season-to-Date Average', fontsize=12)
    plt.ylabel('Test Accuracy', fontsize=12)
    plt.xticks(range(1, 22))
    plt.grid(True)
    plt.legend(title='Test Season', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

all_preds_df = pd.concat(all_svr_predictions, ignore_index=True)
all_preds_df.to_csv("all_regression_predictions_svr_train2.csv", index=False)
print("✅ All per-game regression predictions saved: all_regression_predictions_svr_train2.csv")
```
